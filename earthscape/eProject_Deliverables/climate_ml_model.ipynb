# EarthScape Climate Data - Anomaly Detection Model
This notebook demonstrates the Machine Learning component of the EarthScape Climate Intelligence Platform. We use an Isolation Forest algorithm to detect anomalies in climate data (temperature and CO2 levels).

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest
import matplotlib.pyplot as plt
import seaborn as sns

# Simulate fetching data from Hadoop HDFS / MongoDB
np.random.seed(42)
data_size = 1000
temperatures = np.random.normal(loc=15.0, scale=5.0, size=data_size)
co2_levels = np.random.normal(loc=400.0, scale=20.0, size=data_size)

# Inject some anomalies (extreme weather events)
temperatures[::50] += np.random.normal(loc=20.0, scale=5.0, size=20)
co2_levels[::50] += np.random.normal(loc=100.0, scale=30.0, size=20)

df = pd.DataFrame({'Temperature_C': temperatures, 'CO2_ppm': co2_levels})
df.head()
```

```python
# Initialize and train the Isolation Forest model
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(df)

# Predict anomalies (-1 indicates anomaly, 1 indicates normal)
df['Anomaly'] = model.predict(df[['Temperature_C', 'CO2_ppm']])

print("Total Anomalies Detected:", (df['Anomaly'] == -1).sum())
```

```python
# Visualize the results
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Temperature_C', y='CO2_ppm', hue='Anomaly', palette={1: 'blue', -1: 'red'})
plt.title('Climate Data Anomaly Detection (Isolation Forest)')
plt.xlabel('Temperature (°C)')
plt.ylabel('CO2 Levels (ppm)')
plt.show()
```
